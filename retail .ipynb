## UniVariate Analysis

##### **BAR PLOT =  A bar plot is a graphical representation of data where rectangular bars are used to compare different categories. The length or height of each bar is proportional to the values they represent. Bar plots can be created vertically or horizontally.**

```python
plt.figure(figsize = (5,3))
colors = ['brown', 'green']
plt.figure(figsize = (4,4))
df['Gender'].value_counts().plot(kind= 'bar',edgecolor = 'black',color = colors)
plt.title('Frequency distribution of the Gender')
plt.xlabel("Gender") 
plt.show()
```

### Female customers are doing more retail shopping compare to male customers.

##### **HISTOGRAM PLOT = A histogram plot is a graphical representation of the distribution of numerical data. It groups the data into bins (intervals) and displays the frequency of data points in each bin as rectangular bars.**

```python
plt.figure(figsize = (5,3))
plt.figure(figsize=(5,5))
df['Age'].plot(kind='hist',edgecolor='black')
plt.title('Frequency distribution of the Age')
plt.xlabel("Age") 
plt.show()
```

### From the given data the customers above age 60 are doing more retail shopping and the customers of age 40 are doing less shopping.


##### **PIE PLOT = A pie plot is a circular statistical graphic divided into slices to illustrate numerical proportions. Each slice's size represents its proportion of the total.**

```python
plt.figure(figsize = (5,3))
df['Product Category'].value_counts().plot(kind='pie',autopct="%.2f%%")
plt.title("Pie chart of Product Category")
plt.show()
```

### 35% of the people are doing the clothing shopping and 34 percent of people are doing electronics shopping and the beauty sales is 30 percent.


```python
plt.figure(figsize = (5,3))
colors = ['green', 'pink','purple','blue']
plt.figure(figsize = (5,5))
df['Quantity'].value_counts().sort_index().plot(kind="bar",color = colors,edgecolor = 'black')
plt.title('Frequency distribution of the Quantity')
plt.xlabel("Quantity") 
plt.show()
```

### Most of the people are buying 4 quantities.

```python
plt.figure(figsize = (5,3))
colors=['brown','lightgreen','blue','yellow','pink']
plt.figure(figsize=(5,5))
df['Price per Unit'].value_counts().sort_index().plot(kind='bar',color=colors,edgecolor='black')
plt.title("Frequency distribution of the Price per unit")
plt.show()
```

### PIVOT PLOT = A pivot plot is a visualization built on top of a pivot table, typically created in Python using libraries like pandas and matplotlib or seaborn. It allows you to summarize and plot data by rearranging and aggregating values based on specific rows and columns.

### Retail Dataset : Analyze seasonal demand patterns for products

```python
monthly_product_sales = df.groupby(['Month', 'Product Category'])['Total Amount'].sum().unstack()
print(monthly_product_sales)
```

```python
monthly_product_sales.plot(kind='bar')
plt.title('Monthly Sales Comparison by Product')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```

### lower sales indicate when products are in low demand, and businesses can plan for inventory reduction or promotions.


### High Season Month March might show high sales due to holiday shopping. This indicates seasonal demand for specific products.

### Identify top-performing products or regions

```python
df.head()
```

```python
top_products_revenue = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
print("Top 10 Products by Revenue:")
print(top_products_revenue.head(10))
```

```python
plt.figure(figsize=(8, 3))
top_products_revenue.head(10).plot(kind='bar', color='skyblue')
plt.title('Top  Products by Revenue')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.xticks(rotation = 90)
plt.show()
```

### To determine top-performing products, we need to calculate total revenue or quantity sold for each product.
### Focus on high-revenue products with a strong profit margin.
### Understand seasonal patterns for the top-performing items.

```python
top_products_quantity = df.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False)
print("Top  Products by Quantity Sold:")
print(top_products_quantity)
```

```python
plt.figure(figsize=(5, 3))
top_products_quantity.head().plot(kind='bar')
plt.title('Top Products by Quantity Sold')
plt.xlabel('Product', fontsize=12)
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```
